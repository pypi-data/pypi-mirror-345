```python
import pandas as pd

df = pd.read_csv("datasets/catalog/fairface/labels.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder

columns_to_encode = ["age", "gender", "race", "service_test"]

mapping = {}

for column in columns_to_encode:
    df[column] = df[column].astype(str)
    label_encoder = LabelEncoder()

    df[column] = label_encoder.fit_transform(df[column].astype(str))

    mapping[column] = dict(
        zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))
    )

df
```

```python
reverse_mapping = {
    str(column): {int(value): str(key) for key, value in mapping[column].items()}
    for column in mapping
}

# Convert keys to int, columns to string and values to string
# reverse_mapping = {
#     str(column): {str(value): str(key) for key, value in reverse_mapping[column].items()}
#     for column in reverse_mapping
# }

# to yaml
import yaml

with open("datasets/catalog/fairface/labels_mapping.yaml", "w") as f:
    yaml.dump(reverse_mapping, f)
```

```python
# Check dtypes in reverse mapping
for column in reverse_mapping:
    it, vt = reverse_mapping[column].items(), mapping[column].items()
    print(
        column,
        all([isinstance(k, str) for k, v in it]),
        all([isinstance(k, int) for k, v in vt]),
    )
```
