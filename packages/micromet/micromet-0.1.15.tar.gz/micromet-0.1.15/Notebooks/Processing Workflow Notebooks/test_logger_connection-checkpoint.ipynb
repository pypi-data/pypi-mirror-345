Patch into the Google Cloud Database using the PostGREST API. 

```python
station = 'US-UTW'
startdate = '2024-01-01'

headdict = {'Accept-Profile': 'groundwater','Content-Type': 'application/json'}
params = {'stationid':f'eq.{station}',
         'datetime_start':f'gte.{startdate}'
         }
resp = requests.get("https://ugs-koop-umfdxaxiyq-wm.a.run.app/amfluxeddy",headers=headdict,params=params)
```

```python
data = pd.DataFrame(resp.json())
data['datetime_start'] = pd.to_datetime(data['datetime_start'])
data = data.set_index(['stationid','datetime_start'])
data.loc[station,'co2'].replace(-9999,np.nan).plot()
```
